```python
import numpy as np
import matplotlib.pyplot as plt
#https://github.com/millennial-geoscience/Common_Ss_geodata_examples/blob/main/synthetic_listric_fault_cube_128x128x128.npy
#wget 
data = np.load('/content/synthetic_listric_fault_cube_128x128x128.npz.npy')
plt.imshow(data[20,:,:].T)
```

```python
a = data.transpose(0,1,2)
store = np.ones((a.shape[0],a.shape[1]))
a.shape[0]
for x in range(a.shape[0]):
  for y in range(a.shape[1]):
    trace = a[x,y,:]
    val = np.max(trace)
    idx = list(trace).index(val)
    store[x,y] = idx

plt.imshow(store)
```
